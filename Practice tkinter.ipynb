```python
from tkinter import *

windows = Tk()
windows.title("Simple Calculator")
windows.geometry("400x300")
windows.eval('tk::PlaceWindow . center')

num1 = Entry(windows)
num2 = Entry(windows)

result_lbl = Label(windows, text='Result')

def add():
    result = int(num1.get()) + int(num2.get())
    result_lbl.config(text=result)
    
def subtract():
    result = int(num1.get()) - int(num2.get())
    result_lbl.config(text=result)
    
def multiply():
    result = int(num1.get()) * int(num2.get())
    result_lbl.config(text=result)

def divide():
    result = int(num1.get()) / int(num2.get())
    result_lbl.config(text=result)


add_btn = Button(windows, text='+', width=10, command=add)
minus_btn = Button(windows, text='-', width=10, command=subtract)
multiply_btn = Button(windows, text='*', width=10, command=multiply)
divide_btn = Button(windows, text='/', width=10, command=divide)

num1.pack()
num2.pack()
add_btn.pack()
minus_btn.pack()
multiply_btn.pack()
divide_btn.pack()
result_lbl.pack()

windows.mainloop()
```

```python
from tkinter import *

windows = Tk()
windows.title("My intro")

intro_lbl = Label(windows, text="My introduction")
firstname_lbl = Label(windows, text="First name: ")
lastname_lbl = Label(windows, text="Last name: ")

firstname = Entry(windows)
lastname = Entry(windows)

def submit():
    content = "My name is " + firstname.get() + " " + lastname.get()
    intro_lbl.config(text=content)

submit_btn = Button(windows, text="Submit", command=submit)

intro_lbl.grid(row=0, column=0, columnspan=2)
firstname_lbl.grid(row=1, column=0)
lastname_lbl.grid(row=2, column=0)

firstname.grid(row=1, column=1)
lastname.grid(row=2, column=1)
submit_btn.grid(row=3, column=1)

windows.mainloop()
```

```python
from tkinter import *
from tkinter import messagebox

windows = Tk()
windows.title("BMI Calculator")

intro_lbl = Label(windows, text="BMI Calculator")
myweight_lbl = Label(windows, text="My weight(kg)")
myheight_lbl = Label(windows, text = "My height(cm)")

myweight = Entry(windows)
myheight = Entry(windows)

def submit():
    weight = float(myweight.get())
    height = float(myheight.get())
    bmi = int(weight/(height*height))
    messagebox.showinfo("BMI Result", f"Your BMI is {bmi}") 
    
submit_btn = Button(windows,text="Submit", command = submit)

intro_lbl.grid(row=0,column=0,columnspan=2)
myweight_lbl.grid(row=1,column=0)
myheight_lbl.grid(row=2,column=0)

myweight.grid(row=1,column=1)
myheight.grid(row=2,column=1)
submit_btn.grid(row=3, column=1)

windows.mainloop()
```
