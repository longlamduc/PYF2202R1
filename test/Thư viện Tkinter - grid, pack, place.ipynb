```python
# Setup windows
from tkinter import *

windows = Tk()
windows.geometry("400x300")

message_label = Label(windows, text="Please enter email and password to login")

username_label = Label(windows, text="Username")

password_label = Label(windows, text="Password")

username_entry = Entry(windows)

password_entry = Entry(windows)

login_button = Button(windows, text="Login", command=None)

add_button = Button(windows, text= "Add", command = None, height= 1, width = 10)

delete_button = Button(windows, text= "Delete", command = None, height= 1, width = 10)

reset_button = Button(windows, text= "Reset", command = None, height= 1, width = 10)

search_button = Button(windows, text= "Search", command = None, height= 1, width = 10)

submit_button = Button(windows, text= "Submit", command = None, height= 1, width = 10)
```

## pack

```python
message_label.pack()
username_label.pack()
username_entry.pack()
password_label.pack()
password_entry.pack()
login_button.pack()
```

### Kết quả:


## grid

```python
message_label.grid(column=1, row=0)
username_label.grid(column=0, row=1)
username_entry.grid(column=1, row=1)
password_label.grid(column=0, row=2)
password_entry.grid(column=1, row=2)
login_button.grid(column=1, row=3)
```

### Kết quả:


## place

```python
message_label.place(x=80, y=10)
username_label.place(x=5, y=50)
password_label.place(x=5, y=80)
username_entry.place(x=70, y=50)
password_entry.place(x=70, y=80)
login_button.place(x=100, y=120)
```

### Kết quả:


```python
# mainloop
windows.mainloop()
```
